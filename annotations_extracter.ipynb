```python
import os
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
import mne
```

```python
annotations_a = r'A:\Sano\neonatal_data\annotations_2017_A_fixed.csv'
annotations_b = r'A:\Sano\neonatal_data\annotations_2017_B.csv'
annotations_c = r'A:\Sano\neonatal_data\annotations_2017_C.csv'
```

```python
annot_a = pd.read_csv(annotations_a).transpose().to_numpy()
annot_b = pd.read_csv(annotations_b).transpose().to_numpy()
annot_c = pd.read_csv(annotations_c).transpose().to_numpy()
unified_annot = np.empty((annot_a.shape[0],annot_a.shape[1]))
unified_annot[:] = np.NaN
```

```python
nan = np.NaN
patient_count = annot_a.shape[0]
second_count = annot_a.shape[1]
for n in range(0,patient_count): 
    for s in range (0,second_count):
        #print(annot_a[n][s])
        if (annot_a[n][s] == 1) and (annot_b[n][s] == 1)  and (annot_c[n][s] == 1):
            unified_annot[n][s] = 1
        elif  math.isnan(annot_a[n][s]) and math.isnan(annot_b[n][s]) and math.isnan(annot_c[n][s]):
            unified_annot[n][s] = nan
        else:
            unified_annot[n][s] = 0
```

```python
new_df = pd.DataFrame(unified_annot)
new_df.index +=1
save_df = new_df.transpose()
```

```python
save_df.to_csv(r'seizure_events\consensus_annotations.csv',index=False)
```
