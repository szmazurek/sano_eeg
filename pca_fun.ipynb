```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
sine1 = np.expand_dims(np.sin(3*np.linspace(0,10,100)+2),axis=0) + np.random.normal(0, 1, [1,100])
sine2 = np.expand_dims(np.sin(3*np.linspace(0,10,100)+3),axis=0) + np.random.normal(0, 3, [1,100])

sine =  np.concatenate([sine1,sine2],axis=0)
```

```python
plt.plot(sine1[0])
plt.show()
```

```python
sine.shape
```

```python
pca = PCA()
pca.fit(sine)
```

```python
components = pca.components_
mu = pca.mean_
eigenvals = np.expand_dims(pca.singular_values_,axis=1)
transform_sine1 = pca.transform(np.expand_dims(sine[0],axis=0))
```

```python
print("transofmration", transform_sine1.shape)
print("components", components.shape)
```

```python
reconstructed1 = mu + transform_sine1 @ components
#reconstructed1 = mu + transform_sine1[0,0] * components[0]
```

```python
pca.explained_variance_
```

```python
np.sum(reconstructed1-sine1)
```

```python
a = plt.plot(reconstructed1)
b = plt.plot(sine1[0])
plt.show()
```
